## 1. Data Cleaning and Feature Extraction

```python
# Imputation: From Describe analysis (Cell 5) we can claim the data is fairly clean and need only a few null imputation based on logical assumptions

df['activated_date'] = df['activated_date'].fillna(df['activated_date'].mode()[0])
df['last_payment_date'] = df['last_payment_date'].fillna(df['last_payment_date'].mode()[0])

df['cash_advance'] = df['cash_advance'].fillna(0)
df['minimum_payments'] = df['minimum_payments'].fillna(0)

#only a few rows but still worths the conversion
df['balance'] = df['balance'].fillna(df['balance'].mean()) 
df['credit_limit'] = df['credit_limit'].fillna(df['credit_limit'].mean())
```

```python
# Feature Extraction would include time and some logical KPIs
# Time
df['activated_days'] = (pd.to_datetime('2020-07-31') - df['activated_date']).dt.days #2020-07-31 is the maximum for the activated_date
df['lst_pmt_days'] = (pd.to_datetime('2020-10-18') - df['last_payment_date']).dt.days #2020-10-18 is the max for the last_payment_date

# KPIs
df_report['purchases']=(df_report['purchases']/df_report['credit_limit'])
df_report['cash_advance_pct']=(df_report['cash_advance']/df_report['credit_limit'])
# There could be more KPIs depending on the context and business priorities
```

## Shaping Train and Test datasets

```python
from sklearn.model_selection import train_test_split

X = np.array(df.iloc[:, df.columns != 'fraud'])
y = np.array(df.iloc[:, df.columns == 'fraud'])
print('Shape of X: {}'.format(X.shape))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0, stratify = y) 
# the stratify=y option is a very important option to make sure that we do not get a very unbalanced train vs test in terms of the outcome
print('Shape of X: {}'.format(X_train.shape))
```

```python
# Scaling train data for better accuracy: 
# Because many of the measures they are already between 0-1 then we are using Min Max Scalar 
# Note: y is not getting scaled
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# transform data
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Using SMOTE (Synthetic Minority Oversampling Technique) – Oversampling for Imbalanced Data (Label)

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
X_train_scl_sm,y_train_sm=sm.fit_resample(X_train_scaled, y_train)
```

```python
plt.title('Fraud class histogram after SMOTE oversampling')
plt.xlabel('Fraud')
plt.ylabel('Frequency')
pd.value_counts(y_train_sm).plot.bar();
pd.value_counts(y_train_sm)
```

The "synthetized" training data is now clearly balanced and ready to be trained and get tested

# Modelling 

## Logistic Regression 
we use this as a base model
F1 Score is the main measure to choose the best model as for unbalanced data, accuracy will not be enough

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_recall_curve, auc, roc_auc_score, roc_curve, recall_score, classification_report
import warnings
warnings.filterwarnings(action="ignore")

#Hyper parameter optimization
parameters = {
    'C': np.linspace(1, 8, 8)
             }
lr = LogisticRegression()
clf = GridSearchCV(lr, parameters, cv=5, verbose=5, n_jobs=3)
clf.fit(X_train_scl_sm,y_train_sm.ravel())
```

```python
clf.best_params_
```

```python
lr2 = LogisticRegression(C=5,penalty='l2', verbose=5)
lr2.fit(X_train_scl_sm, y_train_sm.ravel())
```

```python
y_pred = lr2.predict(X_test_scaled)
```

## Analysis of the results of the Logistic Regression Model

```python
y_pred = lr2.predict(X_test_scaled)
y_pred = np.squeeze(y_pred)
y_test = np.squeeze(y_test)
print("Confusion Matrix:")
pd.crosstab(np.transpose(y_test), y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
y_score = lr2.fit(X_train_scl_sm, y_train_sm.ravel())

fpr, tpr, thresholds = roc_curve(y_test, y_score.decision_function(X_test_scaled))

roc_auc = auc(fpr,tpr)

# Plot ROC
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b',label='AUC = %0.3f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.0])
plt.ylim([-0.1,1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
roc_auc
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %f' % accuracy)

precision = precision_score(y_test, y_pred)
print('Precision: %f' % precision)

recall = recall_score(y_test, y_pred)
print('Recall: %f' % recall)

f1 = f1_score(y_test, y_pred)
print('F1 score: %f' % f1)
```

F1 Score of 0.31 is our base score Note that for imbalanced data, Accuracy measure is not reliable so 97% is not easily interpretable

## XGBoost
more advanced models are getting compared

```python
#importing libraries, definitions, training, and tuning (mild tuning due to time and computational power limitation )
from xgboost import XGBClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn import metrics 
from sklearn.model_selection import GridSearchCV

parameters = {
            'max_depth':range(3,10,2),
            'min_child_weight':range(1,5,1)
             }

gsearch1 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=5,
 min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27), 
 param_grid = parameters, scoring='roc_auc',n_jobs=4, cv=5)

gsearch1.fit(X_train_scl_sm, y_train_sm);
```

```python
gsearch1.best_params_, gsearch1.best_score_
```

```python
# Another round of optimization
parameters2 = {
            'max_depth':[8,9,10],
            'min_child_weight':[1,2,3]
             }

gsearch1 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=9,
 min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27), 
 param_grid = parameters2, scoring='roc_auc',n_jobs=4, cv=5)

gsearch1.fit(X_train_scl_sm, y_train_sm);
```

```python
gsearch1.best_params_, gsearch1.best_score_
```

```python
# Tune gamma
# Another round of optimization
parameters3 = {
            'gamma':[i/10.0 for i in range(0,5)]
             }

gsearch1 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=8,
 min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27), 
 param_grid = parameters3, scoring='roc_auc',n_jobs=4, cv=5)

gsearch1.fit(X_train_scl_sm, y_train_sm);
```

```python
gsearch1.best_params_, gsearch1.best_score_
```

```python
# Creating Final Model
# The gamma paramete shall be changed to 0.2

xgb = XGBClassifier(
 learning_rate =0.1,
 n_estimators=1000,
 max_depth=8,
 min_child_weight=1,
 gamma=0.2,
 subsample=0.8,
 colsample_bytree=0.8,
 objective= 'binary:logistic',
 nthread=4,
 scale_pos_weight=1,
 seed=27)

xgb.fit(X_train_scl_sm, y_train_sm,  eval_metric="auc",verbose=False);
```

## Analysis of the results of the XGBoos Classifier Model

```python
y_pred = xgb.predict(X_test_scaled)
y_pred = np.squeeze(y_pred)
y_test = np.squeeze(y_test)
print("Confusion Matrix:")
pd.crosstab(np.transpose(y_test), y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
y_pred_prob=xgb.predict_proba(X_test_scaled).T[1]
```

```python


fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

roc_auc = auc(fpr,tpr)

# Plot ROC
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b',label='AUC = %0.3f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.0])
plt.ylim([-0.1,1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %f' % accuracy)

precision = precision_score(y_test, y_pred)
print('Precision: %f' % precision)

recall = recall_score(y_test, y_pred)
print('Recall: %f' % recall)

f1 = f1_score(y_test, y_pred)
print('F1 score: %f' % f1)
```
